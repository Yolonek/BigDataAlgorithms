### Task 3
Find the person with maximal total weight loss.

```python
total_weight_loss = df.wl1 + df.wl2 + df.wl3
person_with_max_loss = total_weight_loss.idxmax()
print(f'Number of person with biggest weight loss: {person_with_max_loss}')
```

### Task 4
Find the person with minimal average self-esteem.

```python
avg_self_esteem = (df.se1 + df.se2 + df.se3) / 3
person_with_min_se = avg_self_esteem.idxmin()
print(f'Number of person with minimal self-esteem: {person_with_min_se}')
```

### Task 5
Sort the rows by weight loss at the second month in descending order.

### Task 6
Print all the rows such that at least one weight loss was higher than 4.

```python
threshold = 4
df[(df.wl1 > threshold) | (df.wl2 > threshold) | (df.wl3 > threshold)]
```
