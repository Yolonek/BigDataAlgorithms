### Task 15
Draw a histogram of the total weight loss for each group.

```python
total_loss = df.wl1 + df.wl2 + df.wl3
new_df = pd.concat([df.group, total_loss], axis=1)
new_df.columns = ['group', 'total_loss']
new_df
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
figure, axes = plt.subplots(3, 1, layout='constrained')
sns.kdeplot(new_df, x='total_loss', hue='group', ax=axes[0])
sns.histplot(new_df, x='total_loss', hue='group', ax=axes[1])
sns.ecdfplot(new_df, x='total_loss', hue='group', ax=axes[2])
figure.set_size_inches(5, 10)
```
